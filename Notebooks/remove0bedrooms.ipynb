```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from utils import *
```

```python
# Load king_country_dataset
df = pd.read_csv('Data/cleaned_house_sales.csv')
df.head()
```

```python
# Load king_country_dataset
base_metrics = pd.read_csv('Metrics/v_.csv')
base_metrics = base_metrics.drop([0, 1, 2, 3], axis=0)
base_metrics
```

## Removing rows with 0 bedrooms

```python
df = df[df['bedrooms']!= 0]
```

## Defining target and features

```python
X = df.drop(['price', 'sqft_lot'], axis=1) # Features
y = df['price'] # Target
```

## Split the data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=13)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(random_state=13)

rf_regressor.fit(X_train, y_train)
```

```python
metrics_df = create_metrics_df()
metrics_df = add_new_metrics(
    metrics_df, 
    rf_regressor,
    X_train,
    y_train,
    "train",
    "Whithout Bedrooms = 0"
)
metrics_df = add_new_metrics(
    metrics_df, 
    rf_regressor,
    X_test,
    y_test,
    "test",
    "Whithout Bedrooms = 0"
)
metrics_df
```

## XGBoost

```python
import xgboost as xgb

xgb_reg = xgb.XGBRegressor()
xgb_reg.fit(X_train, y_train)
```

```python
metrics_df = add_new_metrics(
    metrics_df, 
    xgb_reg,
    X_train,
    y_train,
    "train",
    "Whithout Bedrooms = 0"
)
metrics_df = add_new_metrics(
    metrics_df, 
    xgb_reg,
    X_test,
    y_test,
    "test",
    "Whithout Bedrooms = 0"
)
metrics_df
```
