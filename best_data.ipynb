```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from utils import *
```

```python
# Load king_country_dataset
df = pd.read_csv('Data/v1_house_sales.csv')
df.head()
```

Sheetansh's part

```python
# Add Columns : Age at Sale, Year since renovation, Was Renovated(bool) to the Dataset
df["Age_at_Sale"] = df["year_sold"] - df["yr_built"]

df["Year_since_Renovation"] = df["year_sold"] - df["yr_renovated"]

df.loc[df["yr_renovated"] == 0, "Year_since_Renovation"] = df["year_sold"] - df["yr_built"]

df["was_renovated"] = (df["yr_renovated"] > 0) & (df["yr_renovated"] > df["yr_built"])

df = df.drop(columns = ["yr_built", "year_sold", "yr_renovated"])
```

droping collumns

```python
df = df.drop(columns = ['sqft_lot', 'floors', 'month_sold', 'day_sold', 'bedrooms'])
```

```python
df.to_csv('Data/v3_house_sales.csv')
```
